```
dataset_url = "https://github.com/vishal-pandey/ml-and-ds/blob/master/datasets/Weather_data.csv?raw=true"
```

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```
df = pd.read_csv(dataset_url)
```

```
df.head()
```

```
df.describe()
```

```
df1 = df.iloc[:, [0, 1, 2, 3, 4, 6, 8, 9, 10, 11, 16]]
```

```
df1.head()
```

```
df1 = df1.dropna()
```

```
df1.head()
```

```
print(pd.DataFrame(df1.iloc[:, 1].values)[0].value_counts())
print(pd.DataFrame(df1.iloc[:, 2].values)[0].value_counts())
print(pd.DataFrame(df1.iloc[:, 3].values)[0].value_counts())
print(pd.DataFrame(df1.iloc[:, 4].values)[0].value_counts())
print(pd.DataFrame(df1.iloc[:, 5].values)[0].value_counts())
print(pd.DataFrame(df1.iloc[:, 6].values)[0].value_counts())
print(pd.DataFrame(df1.iloc[:, 7].values)[0].value_counts())
print(pd.DataFrame(df1.iloc[:, 8].values)[0].value_counts())
print(pd.DataFrame(df1.iloc[:, 9].values)[0].value_counts())
print(pd.DataFrame(df1.iloc[:, 10].values)[0].value_counts())
```

```
df1['datetime_utc'] = pd.to_datetime(df1['datetime_utc'])
df1.set_index('datetime_utc', inplace= True)
df1 = df1.resample('D').mean()
```

```
df1.head()
```

```
df1[' _tempm'].fillna(df1[' _tempm'].mean(), inplace=True)
df1[' _dewptm'].fillna(df1[' _dewptm'].mean(), inplace=True)
df1[' _fog'].fillna(df1[' _fog'].mean(), inplace=True)
df1[' _hum'].fillna(df1[' _hum'].mean(), inplace=True)
df1[' _hail'].fillna(df1[' _hail'].mean(), inplace=True)
df1[' _pressurem'].fillna(df1[' _pressurem'].mean(), inplace=True)
df1[' _rain'].fillna(df1[' _rain'].mean(), inplace=True)
df1[' _snow'].fillna(df1[' _snow'].mean(), inplace=True)
```

```
df1.to_csv('delhi_weather_simplified.csv')
```
